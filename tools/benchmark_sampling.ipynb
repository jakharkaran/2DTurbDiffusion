### Benchmarking sampling speed for different GPUs
- Conditional diffusion mode
- Ngrid = 512

```python
Ngrid = 64

import numpy as np
import matplotlib.pyplot as plt

batch_size_A100 = [1, 2, 4, 8, 16, 32]
diffusion_steps_per_sec_A100 = [40, 36, 30.5, 19.5, 10.5, 5.5]
time_per_batch_A100 = 1000/np.array(diffusion_steps_per_sec_A100)
time_per_sample_A100 = time_per_batch_A100 / np.array(batch_size_A100)

batch_size_h100 = [1, 2, 4, 8, 16, 32, 64, 128]
diffusion_steps_per_sec_h100 = [54, 59.7, 50, 32.5, 17.5, 9, 4.5, 2.5]
time_per_batch_h100 = 1000/np.array(diffusion_steps_per_sec_h100)
time_per_sample_h100 = time_per_batch_h100 / np.array(batch_size_h100)
```

```python
plt.plot(batch_size_A100, time_per_sample_A100, label='A100', marker='o')
plt.plot(batch_size_h100, time_per_sample_h100, label='H100', marker='o')
plt.xlabel('Batch Size')
plt.ylabel('Time per Sample (s)')
```

## Conclusion
- There is always benefit have to the largest batch size
- The batch size should be atleast 8, batch 16 and plus have small benefits
